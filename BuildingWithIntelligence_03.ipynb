```python
from openai import OpenAI
from config import OPENAI_API_KEY  # import your API key

# Initialize the OpenAI client
client = OpenAI(api_key=OPENAI_API_KEY)

# Define the conversation messages
messages = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content": "Hi! Can you suggest a healthy breakfast recipe?"}
]

# Create a chat completion
response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
    max_tokens=500,  # limit output to 500 tokens
    temperature=0.7  # controls creativity of the response
)

# Print the assistant's reply
print(response.choices[0].message.content)
```

```python
# User asks a follow-up question
messages.append({"role": "user", "content": "Can you make it vegan?"})

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages,
    temperature=0.7
)

# Append the assistant's reply to keep context
assistant_reply = response.choices[0].message.content
messages.append({"role": "assistant", "content": assistant_reply})

print(assistant_reply)
```
